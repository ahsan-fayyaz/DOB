### Create smaller datasets from incidents_df and permits_df (the merged dataset)

```python
#Create smaller datasets
incidents_df_2 = incidents_df[["BIN", "Incident date (MM/DD/YYYY)"]]
permits_df_2 = permits_df[["Bin", "Permit Issued Date", "Permit Expiration Date","Contractor Name","Permit Type","Permit Sub Type"]]

#Make sure BIN column is named the same way in both
permits_df_2 = permits_df_2.rename(columns={'Bin': 'BIN',},)

#Rename the Incident date col
incidents_df_2 = incidents_df_2.rename(columns={'Incident date (MM/DD/YYYY)': 'Incident Date',},)

display(incidents_df_2.head())
display(permits_df_2.head())
```

### Merge the datasets "incidents_df_2" and "permits_df_2"

```python
print(len(incidents_df_2))
merged_dataset = pd.merge(incidents_df_2, permits_df_2, how='left', on="BIN")
print(len(merged_dataset))
merged_dataset.head()
```

### Convert the dates to datetime objects

```python
merged_dataset['Incident Date'] = merged_dataset['Incident Date'].astype('datetime64[ns]')
merged_dataset['Permit Issued Date'] = merged_dataset['Permit Issued Date'].astype('datetime64[ns]')
merged_dataset['Permit Expiration Date'] = merged_dataset['Permit Expiration Date'].astype('datetime64[ns]')
merged_dataset.head()
```

### Create a column for flag

```python
merged_dataset["Flag"] = 0
merged_dataset.head()
```

```python
#display the merged_dataset
merged_dataset
```

### If the incident occured while the permit was active, set flag to 1

```python
for i in range(0,len(merged_dataset)):
    if (merged_dataset["Incident Date"][i] >= merged_dataset["Permit Issued Date"][i] and merged_dataset["Incident Date"][i] <= merged_dataset["Permit Expiration Date"][i]):
        merged_dataset["Flag"][i] = 1

merged_dataset.head()
```

```python
merged_dataset[merged_dataset["BIN"] == "1000060"]
```

### Create a Permit Dictionary

```python
##create dictionary for permits
merged_dataset["BIN"] =merged_dataset["BIN"].astype(str).replace('\.0', '', regex=True)

#copy the dataset into a new df called comp2_t
comp2_t = merged_dataset.copy()

#groupby the BIN and sum all
comp2_t = comp2_t.groupby(['BIN']).sum()

#this is adding an empty string to the end of column name strings
comp2_t = comp2_t.add_suffix('').reset_index()

#sort the BIN values
comp2_t = comp2_t.sort_values(by = 'BIN', ascending=True).reset_index(drop=True)

#make sure the BIN is a str type
comp2_t["BIN"] = comp2_t["BIN"].astype(str)

#set the key:value as BIN:Flag
permitDic = comp2_t.set_index('BIN')['Flag'].to_dict()

# permitDic
```

### Contractor groupby

```python
# Create a new dataframe called con with just two cols
con= merged_dataset[["BIN","Contractor Name","Flag"]]
print(len(con))
con = con[con["Flag"]==1]
con = con.reset_index(drop=True)
print(len(con))
#Drop nan values
con["Contractor Name"] = con["Contractor Name"].astype(str)
con["Contractor Name"] = con["Contractor Name"].astype(str).map(str.strip)


#Drop nan values
con = con[con["Contractor Name"]!='nan']


#drop duplicates code
con = con.drop_duplicates(subset=['BIN','Contractor Name'])

#drop duplicates code
con = con.groupby(['BIN'])['Contractor Name'].apply(', '.join).reset_index()
print(len(con))
con.head(10)
```

### Permits groupby

```python
# Create a new dataframe called permit_type with just two cols
permit_type= merged_dataset[["BIN","Permit Type","Flag"]]
print(len(permit_type))
permit_type = permit_type[permit_type["Flag"]==1]
permit_type = permit_type.reset_index(drop=True)
print(len(permit_type))
#Drop nan values
permit_type["Permit Type"] = permit_type["Permit Type"].astype(str)
permit_type["Permit Type"] = permit_type["Permit Type"].astype(str).map(str.strip)


#Drop nan values
permit_type = permit_type[permit_type["Permit Type"]!='nan']


#drop duplicates code
permit_type = permit_type.drop_duplicates(subset=['BIN','Permit Type'])


#drop duplicates code
permit_type = permit_type.groupby(['BIN'])['Permit Type'].apply(', '.join).reset_index()
print(len(permit_type))
permit_type.head(10)
```

### Permits Subtype groupby

```python
# Create a new dataframe called permit_subtype with just two cols
permit_subtype= merged_dataset[["BIN","Permit Sub Type","Flag"]]
print(len(permit_subtype))
permit_subtype = permit_subtype[permit_subtype["Flag"]==1]
permit_subtype = permit_subtype.reset_index(drop=True)
print(len(permit_subtype))

#Drop nan values
permit_subtype["Permit Sub Type"] = permit_subtype["Permit Sub Type"].astype(str)
permit_subtype["Permit Sub Type"] = permit_subtype["Permit Sub Type"].astype(str).map(str.strip)

#Drop nan values
permit_subtype = permit_subtype[permit_subtype["Permit Sub Type"]!='nan']


#drop duplicates code
permit_subtype = permit_subtype.drop_duplicates(subset=['BIN','Permit Sub Type'])

#Convert to String

#drop duplicates code
permit_subtype = permit_subtype.groupby(['BIN'])['Permit Sub Type'].apply(', '.join).reset_index()
print(len(permit_subtype))
permit_subtype.head(10)
```

```python
# This function will help to evaluate linear relationship for two variables

def evaluate_linear_relationship(a, b):
    slope = np.cov(a, b, bias=True)[0][1] / np.var(a)
    intercept = np.mean(b) - (slope * np.mean(a))
    predictions = (slope * a) + intercept
    residuals = b - predictions
    return slope, intercept, predictions, residuals

# This function will help evaluate the intervals when we reach Confidence Intervals and Predication Intervals part
def get_intervals(a, b, residuals):
    t = stats.t(df=len(a) - 2).ppf(0.975)
    s_err = np.sum(np.power(residuals, 2))
    confidence_interval = t * np.sqrt((s_err/(len(a) - 2))*(1/len(a) + (np.power((a - a.mean()), 2)/((np.sum(np.power(a,2))) - len(full_df)*(np.power(a.mean(),2))))))
    prediction_interval = t * (np.std(residuals))
    return abs(confidence_interval), prediction_interval
```

```python
def missing_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
    '% of Total Values', ascending=False).round(1)
    print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
            " columns that have missing values.")
    return mis_val_table_ren_columns
```
