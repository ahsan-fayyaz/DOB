### Import the DOB Now dataset

```python
DOB_now_permits_df = pd.read_csv(path + "DOBNOW job_permits high risk construction.csv")
DOB_now_permits_df.head()
```

### Cleaning the DOB Now dataset

```python
# Rename the Applicant Business Name to Contractor Name
DOB_now_permits_df = DOB_now_permits_df.rename(columns={'Applicant Business Name': 'Contractor Name',
                                                         'Work Type Name' : 'Permit Sub Type'},)

#Create a Permit Type Column
DOB_now_permits_df["Permit Type"] = "EQ"
```

### Check the value counts of unique work type names and change them to codes instead

```python
DOB_now_permits_df["Permit Sub Type"].value_counts()
```

```python
#Use the codes for Work Type Names

DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Sidewalk Shed','SH')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Plumbing','PL')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Construction Fence','FN')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('General Construction','GC')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Supported Scaffold','SF')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Mechanical Systems','MS')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Sprinklers','SP')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Sign','SG')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Structural','ST')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Boiler Equipment','BE')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Antenna','AN')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Curb Cut','CC')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Standpipe','SD')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Protection and Mechanical Methods','PM')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Foundation','FO')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Earth Work','EA')
DOB_now_permits_df["Permit Sub Type"] = DOB_now_permits_df["Permit Sub Type"].str.replace('Support of Excavation','SE')

# DOB_now_permits_df.head()
```

### Cleaning the BIS dataset

```python
# Rename all cols to match DOB Now dataset

BIS_job_permits_df = BIS_job_permits_df.rename(columns={'Job Type': 'JobType',
                                                        'BIN Number': 'Bin',
                                                        'Borough Name': 'Borough',
                                                        'Job Location House Number': 'House Number',
                                                        'Job Location Street Name': 'Street Name',
                                                        'Applicant Business Name': 'Contractor Name',
                                                        'Proposed Stories': 'ProposeeBuildingStoriesft',
                                                        'Proposed Height': 'ProposedBuildingHeight',
                                                        'Proposed Occupancy Classification Description': 'Proposed Occupancy Classification',
                                                        'D_DATE': 'Permit Issued Date',
                                                        'D_DATE.1': 'Permit Expiration Date',        
                                                        }
                                               ,)

BIS_job_permits_df = BIS_job_permits_df.drop(['Number of Stories'], axis=1)
```
